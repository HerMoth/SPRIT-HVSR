```python
import plotly.graph_objs as go
import ipywidgets as widgets
from IPython.display import display

# Create the Input tab
input_tab = widgets.Textarea(
    value='',
    placeholder='Enter input parameters here...',
    description='Input:',
    disabled=False
)

# Create the Preview tab
preview_fig = go.FigureWidget()
preview_tab = widgets.Output()
with preview_tab:
    display(preview_fig)

# Create the PPSD Settings subtab
ppsd_settings_tab = widgets.Textarea(
    value='',
    placeholder='Enter PPSD settings here...',
    description='PPSD Settings:',
    disabled=False
)

# Create the HVSR Settings subtab
hvsr_settings_tab = widgets.Textarea(
    value='',
    placeholder='Enter HVSR settings here...',
    description='HVSR Settings:',
    disabled=False
)

# Create the Plot Settings subtab
plot_settings_tab = widgets.Textarea(
    value='',
    placeholder='Enter plot settings here...',
    description='Plot Settings:',
    disabled=False
)

# Create the Curve Analysis subtab
curve_fig = go.FigureWidget()
curve_analysis_tab = widgets.Output()
with curve_analysis_tab:
    display(curve_fig)

# Create the Report subtab
report_fig = go.FigureWidget()
report_tab = widgets.Output()
with report_tab:
    display(report_fig)

# Create the Settings tab with subtabs
settings_tab = widgets.Tab()
settings_tab.children = [ppsd_settings_tab, hvsr_settings_tab, plot_settings_tab]
settings_tab.set_title(0, 'PPSD Settings')
settings_tab.set_title(1, 'HVSR Settings')
settings_tab.set_title(2, 'Plot Settings')

# Create the Results tab with subtabs
results_tab = widgets.Tab()
results_tab.children = [curve_analysis_tab, report_tab]
results_tab.set_title(0, 'Curve Analysis')
results_tab.set_title(1, 'Report')

custom_style = {
    'description_width': 'initial',
    'background': 'linear-gradient(to right, #ffafbd, #ffc3a0)',
    'border': 'none',
    'color': '#333333',
    'padding': '10px',
    'font-size': '16px'}

# Create the main tab widget
sprit_widget = widgets.Tab(style=custom_style)
sprit_widget.children = [input_tab, preview_tab, settings_tab, widgets.Output(), results_tab]
sprit_widget.set_title(0, 'Input')
sprit_widget.set_title(1, 'Preview')
sprit_widget.set_title(2, 'Settings')
sprit_widget.set_title(3, 'Log')
sprit_widget.set_title(4, 'Results')


# Display the tab widget
display(sprit_widget)
```
